### <font color="HotPink" > What Is JupyterLite? </font>
JupyterLite is a distribution of JupyterLab. It runs in the browser and is built using JupyterLab components and extensions. JupyterLite is being created to provide a lightweight, easily accessible computing environment. JupyterLite will launch in a web browser and requires no installations of software.
- The JupyterLite project is unofficial and still under development.

JupyterLite is centered upon accessible, browser-based, interactive, computing:

- Python kernel backed by Pyodide running in a Web Worker.
- JavaScript and P5.js kernels running in an IFrame
- Supports visualization libraries including altair, bqplot, ipywidgets, matplotlib, and plotly
- Basic session and kernel management to have multiple kernels running at the same time

Plotly is an interactive, open-source plotting library. 40+ different chart types are supported. They cover things like statistic graphs, 3-D models and scientific graphs.

```python
import plotly.express as px
df = px.data.gapminder().query("year == 2007").query("continent == 'Europe'")
df.loc[df['pop'] < 2.e6, 'country'] = 'Other countries' # Represent only large countries
fig = px.pie(df, values='pop', names='country', title='Population of European continent')
fig.show()
```

### <font color="HotPink" >JupyterLite & GitHub Pages </font>
